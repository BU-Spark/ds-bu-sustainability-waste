```python
import pandas as pd
```

```python
fy22 = pd.read_csv('BU_Daily_Weights_FY22.csv')
fy22[:5]
```

```python
psi_140_bay_state = pd.read_csv('psi_140_bay_state.csv')
psi_140_bay_state[:5]
```

```python
fy22_140_bay_state = fy22[fy22['Address'] == '140 BAY STATE RD']
fy22_140_bay_state[:5]
```

```python
def format_date(df, column):
    return  pd.to_datetime(df[column], format="%Y-%m-%d", infer_datetime_format=True)
```

```python
from datetime import datetime
fy22_140_bay_state['Work Order Date'] = pd.to_datetime(fy22_140_bay_state['Work Order Date'], format="%Y-%m-%d", infer_datetime_format=True)
psi_140_bay_state['date'] = pd.to_datetime(psi_140_bay_state['date'], format="%Y-%m-%d", infer_datetime_format=True)
```

```python
all_140_bay_state = pd.merge(fy22_140_bay_state, psi_140_bay_state, 
                             left_on='Work Order Date', right_on='date', how='inner')
```

```python
all_140_bay_state
```

```python
import statsmodels.api as sm
X = all_140_bay_state[["mean_valuePsi", "mean_fahrenheit"]]
y = all_140_bay_state["Tons"]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X = all_140_bay_state['mean_valuePsi']
X = sm.add_constant(X)
X_train, X_test, y_psi_train, y_psi_test = train_test_split(X, y, test_size=0.3, random_state=42)
model_psi = sm.OLS(y_psi_train, X_train).fit()

y_psi_pred = model_psi.predict(X_test)
mse_psi = np.mean((y_psi_pred - y_psi_test) ** 2)
mse_psi
```

```python
X = all_140_bay_state['mean_fahrenheit']
X = sm.add_constant(X)
X_train, X_test, y_temp_train, y_temp_test = train_test_split(X, y, test_size=0.3, random_state=42)
model_temp = sm.OLS(y_temp_train, X_train).fit()

y_temp_pred = model_temp.predict(X_test)
mse_temp = np.mean((y_temp_pred - y_temp_test) ** 2)
mse_temp
```

```python
def linear_loss_psi(add, psi_add):
    fy22_add = fy22[fy22['Address'] == add]
    fy22_add['Work Order Date'] = pd.to_datetime(fy22_add['Work Order Date'], 
                                                 format="%Y-%m-%d", infer_datetime_format=True, errors = 'coerce')
    psi_add['date'] = pd.to_datetime(psi_add['date'], format="%Y-%m-%d", infer_datetime_format=True)
    all_add = pd.merge(fy22_add, psi_add, 
                             left_on='Work Order Date', right_on='date', how='inner')    
    X = all_add['mean_valuePsi']
    X = sm.add_constant(X)
    y = all_add["Tons"]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model = sm.OLS(y_train, X_train).fit()

    y_pred = model.predict(X_test)
    mse = np.mean((y_pred - y_test) ** 2)
    print(model.summary())
    return mse
```

```python
def linear_loss_temp(add, psi_add):
    fy22_add = fy22[fy22['Address'] == add]
    fy22_add['Work Order Date'] = pd.to_datetime(fy22_add['Work Order Date'], 
                                                 format="%Y-%m-%d", infer_datetime_format=True, errors = 'coerce')
    psi_add['date'] = pd.to_datetime(psi_add['date'], format="%Y-%m-%d", infer_datetime_format=True)
    all_add = pd.merge(fy22_add, psi_add, 
                             left_on='Work Order Date', right_on='date', how='inner')    
    X = all_add['mean_fahrenheit']
    X = sm.add_constant(X)
    y = all_add["Tons"]
    X_train, X_test, y_temp_train, y_temp_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model_temp = sm.OLS(y_temp_train, X_train).fit()

    y_temp_pred = model_temp.predict(X_test)
    mse_temp = np.mean((y_temp_pred - y_temp_test) ** 2)
    print(model_temp.summary())
    return mse_temp
```
