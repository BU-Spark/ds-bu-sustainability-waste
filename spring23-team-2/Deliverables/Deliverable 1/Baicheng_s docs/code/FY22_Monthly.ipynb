```python
def generate_monthly_heatmap(df, month, csv_name, map_name):
    df = df.loc[df['Work Order Date'].
                         apply(lambda x: str(x).split('/')[0]) == month]
    df = df.groupby(['Address', 'Location'], as_index=False)['Tons'].mean()
    df.to_csv(csv_name)
    gps = pd.read_csv('./trash_gps_simplified.csv')
    df = pd.merge(df, gps, on=['Address'])
    base_map = Map()
    df['latitude'] = df['latitude'].astype(float)
    df['longitude'] = df['longitude'].astype(float)
    df['Tons'] = df['Tons'].astype(float)
    mean = df.groupby(['latitude', 'longitude'], as_index=False)['Tons'].mean()
    base_map = Map(location=[42.349396, -71.104056], zoom_start=12)
    HeatMap(mean).add_to(base_map)
    html = base_map.save(map_name)
```

```python
generate_monthly_heatmap(df_trash, '1', 'trash_jan_mean.csv', 'trash_jan_mean.html')
generate_monthly_heatmap(df_trash, '2', 'trash_feb_mean.csv', 'trash_feb_mean.html')
generate_monthly_heatmap(df_trash, '3', 'trash_march_mean.csv', 'trash_march_mean.html')
generate_monthly_heatmap(df_trash, '4', 'trash_april_mean.csv', 'trash_april_mean.html')
generate_monthly_heatmap(df_trash, '5', 'trash_may_mean.csv', 'trash_may_mean.html')
generate_monthly_heatmap(df_trash, '6', 'trash_june_mean.csv', 'trash_june_mean.html')
generate_monthly_heatmap(df_trash, '7', 'trash_july_mean.csv', 'trash_july_mean.html')
generate_monthly_heatmap(df_trash, '8', 'trash_aug_mean.csv', 'trash_aug_mean.html')
generate_monthly_heatmap(df_trash, '9', 'trash_sept_mean.csv', 'trash_sept_mean.html')
generate_monthly_heatmap(df_trash, '10', 'trash_oct_mean.csv', 'trash_oct_mean.html')
generate_monthly_heatmap(df_trash, '11', 'trash_nov_mean.csv', 'trash_nov_mean.html')
generate_monthly_heatmap(df_trash, '12', 'trash_dec_mean.csv', 'trash_dec_mean.html')
```
