```python
import pandas as pd
```

```python
jan = pd.read_csv('./trash_jan_mean.csv')
```

```python
feb = pd.read_csv('./trash_feb_mean.csv')
march = pd.read_csv('./trash_march_mean.csv')
april = pd.read_csv('./trash_april_mean.csv')
may = pd.read_csv('./trash_may_mean.csv')
june = pd.read_csv('./trash_june_mean.csv')
july = pd.read_csv('./trash_july_mean.csv')
aug = pd.read_csv('./trash_aug_mean.csv')
sept = pd.read_csv('./trash_sept_mean.csv')
octo = pd.read_csv('./trash_oct_mean.csv')
nov = pd.read_csv('./trash_nov_mean.csv')
dec = pd.read_csv('./trash_dec_mean.csv')
```

```python
sites_months = {}
def add_to_sites(pd):
    i = 0
    for site in pd['Address'].tolist():
        if site not in sites_months:
            sites_months[site] = []
            sites_months[site].append(pd['Tons'].tolist()[i])
        else:
            sites_months[site].append(pd['Tons'].tolist()[i])
        i += 1
```

```python
add_to_sites(jan)
add_to_sites(feb)
add_to_sites(march)
add_to_sites(april)
add_to_sites(may)
add_to_sites(june)
add_to_sites(july)
add_to_sites(aug)
add_to_sites(sept)
add_to_sites(octo)
add_to_sites(nov)
add_to_sites(dec)
```

```python
def save_plot(key):
    fig = plt.figure()
    if len(sites_months[key]) < 12:
        for i in range(12 - len(sites_months[key])):
            sites_months[key].append(sum(sites_months[key]) / len(sites_months[key]))
    if len(sites_months[key]) > 12:
        for i in range(len(sites_months[key]) - 12):
            del sites_months[key][-1]
    plt.bar(months, sites_months[key])
    plt.xlabel("Months")
    plt.ylabel("Weights of Trash/Tons")
    plt.title(key)
    plt.savefig(key+'.png')
```

```python
for key in list(sites_months.keys()):
    save_plot(key)
```

```python
mean = [jan['Tons'].mean(),
       feb['Tons'].mean(),
       march['Tons'].mean(),
       april['Tons'].mean(),
       may['Tons'].mean(),
       june['Tons'].mean(),
       july['Tons'].mean(),
       aug['Tons'].mean(),
       sept['Tons'].mean(),
       octo['Tons'].mean(),
       nov['Tons'].mean(),
       dec['Tons'].mean()]
```
