```python
import numpy as np
from folium import plugins, Map, Marker, CircleMarker, Icon
from folium.plugins import HeatMap

base_map = Map()
mean_trash['latitude'] = mean_trash['latitude'].astype(float)
mean_trash['longitude'] = mean_trash['longitude'].astype(float)
mean = mean_trash.groupby(['latitude', 'longitude'], as_index=False)['Tons'].mean()
base_map = Map(location=[42.349396, -71.104056], zoom_start=12)
HeatMap(mean).add_to(base_map)
html = base_map.save('trash_mean.html')
base_map
```
