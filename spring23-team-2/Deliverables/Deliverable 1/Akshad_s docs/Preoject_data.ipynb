```python
import pandas as pd
files = pd.read_csv('combined_all_device.csv')
df1 = pd.read_csv('combined_39691_sorted_by_date.csv')
df2 = pd.read_csv('combined_39689_sorted_by_date.csv')
df3 = pd.read_csv('combined_39692_sorted_by_date.csv')
files
```

```python
import matplotlib.pyplot as plt
df = df1.dropna()
dfn = df2.dropna()
dfnn = df3.dropna()
plt.scatter(df['celsius'], df['valuePsi'])
plt.scatter(dfn['celsius'], dfn['valuePsi'])
plt.scatter(dfnn['celsius'], dfnn['valuePsi'])
plt.xlabel('Temp (C)')
plt.ylabel('PSI')
plt.title('Temperature in Celsius vs. PSI')
plt.tight_layout()  
plt.show()
```

```python
plt.plot(dfn['readingDtUtz'], dfn['valuePsi'])
plt.plot(dfn['readingDtUtz'], dfn['celsius'])
plt.tight_layout()
plt.show()
```

```python
r = df1.iloc[100:180]
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)
plt.plot(r['readingDtUtz'], r['valuePsi']/10)
plt.plot(r['readingDtUtz'], r['celsius'])
plt.xticks(rotation=90)
plt.show()
```
